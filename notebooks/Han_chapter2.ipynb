### Nominal attributes
The values of a **nominal attribute** are symbols or names of things, nominal attributes are also referred to as **categorical**. The values do not have any meaningful order.

```python
import pandas as pd
data = {
    'Names': ['ali', 'sina', 'mohammad'],
    'Cities': ['Tehran', 'Shiraz', 'Rasht'],
    'Universities': ['IUST', 'SUT', 'SBU']
}

df = pd.DataFrame(data)
df.head()
```

### Binary attributes
A **binary attribute** is a nominal attribute with only two categories or states: 0 or 1, where 0 typically
means that the attribute is absent, and 1 means that it is present. Binary attributes are referred to as
Boolean if the two states correspond to *true* and *false*.

```python
data = {
    'courses': ['data-mining', 'data-structures', 'ml-theory'],
    'is_offered_this_term': [1, 0, 1]
}
df = pd.DataFrame(data)
df.head()
```

### Ordinal attributes
An **ordinal attribute** is an attribute with possible values that have a meaningful order or ranking among
them, but the magnitude between successive values is not known.

```python
data = {
    'students': ['rahimi', "jamshidi", "rezaei"],
    'course_overview': ['didn\'t pass', 'not bad', 'well']
}

df = pd.DataFrame(data)
df.head()
```

### numeric attributes
A **numeric attribute** is quantitative; it is a measurable quantity, represented in integer or real values. Numeric attributes can be interval-scaled or ratio-scaled.

```python
data = {
    'product' :['phone', 'laptop', 'mouse'],
    'price_in_dollars': [999, 1499, 99],
    'discount_in_percent': [5, 7, 2]
}
df = pd.DataFrame(data)

print(df.describe())
df['normalized_price'] = (df['price_in_dollars'] - df['price_in_dollars'].mean()) / df['price_in_dollars'].std()
df.head()
```

### Discrete vs Continous attribute
A **discrete attribute** has a finite or countably infinite set of values, which may or may not be represented as integers.

If an attribute is not discrete, it is **continuous**.

```python
import random
data = {
    'classes_size' : [100, 50, 60, 30], # discrete
    'student_heights': [i for i in range(170, 199)]  # continous
}
df = pd.DataFrame(data['classes_size'])
df['student_heights'] =  data['student_heights'][:4]
df.head()
```

### measuring centeral tendency


*   **mean** = sum of observes / number of observes
*   **median** = middle value in a set of ordered data values
*   **mode** = value that occurs most frequently compared to all neighboring values in the set.
*   **midrange** = average of the largest and smallest values in the set.




```python
import numpy as np
salary_data_in_dollars_monthly = np.array([1000, 1500, 900, 750, 500, 2000, 1500, 1200, 1100, 450])

# mean
mean = np.sum(salary_data_in_dollars_monthly) / np.size(salary_data_in_dollars_monthly)

# median
salary_data_in_dollars_monthly.sort()
median = salary_data_in_dollars_monthly[np.size(salary_data_in_dollars_monthly) // 2]

# mode
dict = {}
for item in salary_data_in_dollars_monthly:
  if item in dict:
    dict[item] += 1
  else:
    dict.update({item : 1})
mode = max(zip(dict.values(), dict.keys()))[1]

# midrange
mid_range = (salary_data_in_dollars_monthly[0] + salary_data_in_dollars_monthly[np.size(salary_data_in_dollars_monthly) - 1]) / 2

print(f"mean of data is {mean}")
print(f"median of data is {median}")
print(f"mode of data is {mode}")
print(f"mid range of data is {mid_range}")

''' by using numpy and scipy :
import numpy as np
import scipy
  mean = np.mean(array)
  median = np.median(array)
  mode = scipy.stats.mode(arr)
  '''
```

### measuring dipersion of data

*   Range, quartiles
*   boxplot, outliers
*   variance and standard deviation



```python
# range and quartiles
import numpy as np

arr = np.array([10, 4, 23, 54, 79, 9, 67, 35, 3, 64])

range = np.ptp(arr) # difference maximum between minimum of data
print(f"range is: {range}")

quantile = np.quantile(arr, 0.25) # computes the q-th quantile of data along the specified axis
print(f"quantile is: {quantile}")
```

```python
# boxplot and outliers

import matplotlib.pyplot as plt

np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(2, 3))
plt.boxplot(data)
plt.show()
```

```python
# varince and standard deviation

import numpy as np
arr = np.array([10, 4, 23, 54, 79, 9, 67, 35, 3, 64])

var = np.var(arr)
std = np.std(arr)
print(f"variance of data is :{var} and standard deviation is: {std}")
```

### covariance and corelation analysis


*   covariance of numeric data
*   correlation coefficient for numeric data

$covariance(A,B) = E((A - A̅) (B - B̅)) = ((Σ((a_i - A̅)(b_i - B̅))) / n)$

```python
import numpy as np
a = np.array([12, 34, 46, 43, 32, 17, 65, 76])
b = np. array([45, 34, 32, 54, 67, 87, 45, 54])

cov_matrix = np.cov(a, b)
cov_matrix
```
