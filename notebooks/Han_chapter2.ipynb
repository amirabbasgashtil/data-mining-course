### Nominal attributes
The values of a **nominal attribute** are symbols or names of things, nominal attributes are also referred to as **categorical**. The values do not have any meaningful order.

```python
import pandas as pd
data = {
    'Names': ['ali', 'sina', 'mohammad'],
    'Cities': ['Tehran', 'Shiraz', 'Rasht'],
    'Universities': ['IUST', 'SUT', 'SBU']
}

df = pd.DataFrame(data)
df.head()
```

### Binary attributes
A **binary attribute** is a nominal attribute with only two categories or states: 0 or 1, where 0 typically
means that the attribute is absent, and 1 means that it is present. Binary attributes are referred to as
Boolean if the two states correspond to *true* and *false*.

```python
data = {
    'courses': ['data-mining', 'data-structures', 'ml-theory'],
    'is_offered_this_term': [1, 0, 1]
}
df = pd.DataFrame(data)
df.head()
```

### Ordinal attributes
An **ordinal attribute** is an attribute with possible values that have a meaningful order or ranking among
them, but the magnitude between successive values is not known.

```python
data = {
    'students': ['rahimi', "jamshidi", "rezaei"],
    'course_overview': ['didn\'t pass', 'not bad', 'well']
}

df = pd.DataFrame(data)
df.head()
```

### numeric attributes
A **numeric attribute** is quantitative; it is a measurable quantity, represented in integer or real values. Numeric attributes can be interval-scaled or ratio-scaled.

```python
data = {
    'product' :['phone', 'laptop', 'mouse'],
    'price_in_dollars': [999, 1499, 99],
    'discount_in_percent': [5, 7, 2]
}
df = pd.DataFrame(data)

print(df.describe())
df['normalized_price'] = (df['price_in_dollars'] - df['price_in_dollars'].mean()) / df['price_in_dollars'].std()
df.head()
```

### Discrete vs Continous attribute
A **discrete attribute** has a finite or countably infinite set of values, which may or may not be represented as integers.

If an attribute is not discrete, it is **continuous**.

```python
import random
data = {
    'classes_size' : [100, 50, 60, 30], # discrete
    'student_heights': [i for i in range(170, 199)]  # continous
}
df = pd.DataFrame(data['classes_size'])
df['student_heights'] =  data['student_heights'][:4]
df.head()
```

### measuring centeral tendency


*   **mean** = sum of observes / number of observes
*   **median** = middle value in a set of ordered data values
*   **mode** = value that occurs most frequently compared to all neighboring values in the set.
*   **midrange** = average of the largest and smallest values in the set.




```python
import numpy as np
salary_data_in_dollars_monthly = np.array([1000, 1500, 900, 750, 500, 2000, 1500, 1200, 1100, 450])

# mean
mean = np.sum(salary_data_in_dollars_monthly) / np.size(salary_data_in_dollars_monthly)

# median
salary_data_in_dollars_monthly.sort()
median = salary_data_in_dollars_monthly[np.size(salary_data_in_dollars_monthly) // 2]

# mode
dict = {}
for item in salary_data_in_dollars_monthly:
  if item in dict:
    dict[item] += 1
  else:
    dict.update({item : 1})
mode = max(zip(dict.values(), dict.keys()))[1]

# midrange
mid_range = (salary_data_in_dollars_monthly[0] + salary_data_in_dollars_monthly[np.size(salary_data_in_dollars_monthly) - 1]) / 2

print(f"mean of data is {mean}")
print(f"median of data is {median}")
print(f"mode of data is {mode}")
print(f"mid range of data is {mid_range}")

''' by using numpy and scipy :
import numpy as np
import scipy
  mean = np.mean(array)
  median = np.median(array)
  mode = scipy.stats.mode(arr)
  '''
```

### Data matrix vs dissimilarity matrix
**Data matrix** stores the n data objects in the form
of a relational table or an n-by-p matrix (n objects * p attributes).

**dissimilarity matrix** stores a collection of proximities
that are available for all pairs of n objects. It is often represented by an n-by-n table.

d(i, j) is the measured dissimilarity or “difference” between objects i and j . In general,
d(i, j) is a nonnegative number that is close to 0 when objects i and j are highly similar or “near”
each other, and becomes larger the more they differ.

```python
# data matrix
import pandas as pd

data_matrix = {
    "feature1" : [10, 12, 34 ,45, 54],
    "feature2" : [10, 20, 30, 40, 50]
}
# now number of features = 2 and number of observations = 5
# data matrix is 5*2
df = pd.DataFrame(data_matrix)
df
```

```python
# dissimilarity matrix
from scipy.spatial.distance import pdist, squareform
import pandas as pd

data_matrix = {
    'Feature1': [10, 12, 34, 45, 54],
    'Feature2': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data_matrix)

dissimilarity_matrix = squareform(pdist(df[['Feature1', 'Feature2']], metric='euclidean')) # euclidean distance

df_ = pd.DataFrame(dissimilarity_matrix)
df_
```

### Proximity measures for nominal attributes
The dissimilarity
between two objects i and j can be computed based on the ratio of mismatches

```python
from sklearn.metrics import jaccard_score
import matplotlib.pyplot as plt

data = {
    'Instance1': ['Red', 'Square', 'Large', 'dog'],
    'Instance2': ['Blue', 'Circle', 'Small', 'cat']
}

# Convert nominal attributes to binary values (pd.get_dummies)
binary_data = pd.get_dummies(pd.DataFrame(data))
print(binary_data)
jaccard_similarity = jaccard_score(binary_data.iloc[0], binary_data.iloc[1])
jaccard_similarity_1 = jaccard_score(binary_data.iloc[0], binary_data.iloc[0])

for i in range(len(binary_data)):
  print(binary_data.iloc[i], end="\n")

print(f"Jaccard Similarity: {jaccard_similarity}")
print(f"Jaccard Similarity: {jaccard_similarity_1}")
```

### Dissimilarity of numeric data: Minkowski distance
d(v, u)= $1\over (Σ(v_i - u_i)^p)^p$

```python
from scipy.spatial.distance import minkowski
v = [1, 3, 5, 5, 7, 8]
u = [4, 5, 3, 5, 7, 2]

distance = minkowski(u, v, p=2)
distance
```

### Dissimilarity for attributes of mixed types

suppose that we have different data types in our data base for measuring dissimilarity what we can do? one solution is gower's distance

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler


data = {
    'Numeric': [25, 19, 23, 4, 5],
    'Binary': [1, 0, 1, 1, 0],
    'Nominal': ['A', 'B', 'C', 'D', 'E'],
    'Ordinal': [1, 3, 2, 4, 5]
}


numeric_data = np.array([data['Numeric']]).reshape(-1, 1)
numeric_data = StandardScaler().fit_transform(numeric_data)

binary_data = pd.get_dummies(pd.DataFrame(data['Binary'], columns=['Binary']))

nominal_data = pd.get_dummies(pd.DataFrame(data['Nominal'], columns=['Nominal']))


mixed_data = np.concatenate((numeric_data, binary_data, nominal_data, np.array([data['Ordinal']]).T), axis=1)

distances_oned = pdist(mixed_data, metric='euclidean')
print(f"distances in 1d array:\n{distances_oned}")
distance_sqrd = squareform(distances_oned)
print(f"distances in square form:\n{distance_sqrd}")
```

### cosine similarity
Measures the cosine of the angle between two binary vectors.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "apple is a fruit",
    "orange is orange",
    "Python programming is widely used in data science applications."
]

vectorizer = CountVectorizer()
document_matrix = vectorizer.fit_transform(documents)
print(f"doc matrix:\n{document_matrix}")
cosine_similarities = cosine_similarity(document_matrix, document_matrix)
print("Cosine Similarity Matrix:")
print(cosine_similarities)
```

### Measuring similar distributions: the Kullback-Leibler divergence

```python
import numpy as np
from scipy.stats import entropy
import matplotlib.pyplot as plt


distribution1 = np.random.normal(0.25, 0.1, 100)
distribution2 = np.random.normal(0.4, 0.2, 100)

plt.figure(figsize=(3,3))
bins = np.linspace(-0.2, 0.5, 50)
plt.hist(distribution1, bins, alpha=0.7, color='blue')
plt.hist(distribution2, bins, alpha=0.7, color='green')
plt.show()
kl_divergence = entropy(distribution1, distribution2)
print(f"Kullback-Leibler Divergence: {kl_divergence}")
```

###  Capturing hidden semantics in similarity measures

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Example dataset with text documents
documents = [
    "Data mining is an exciting field of study.",
    "Machine learning techniques are used for data analysis.",
    "Python programming is widely used in data science applications.",
    "python and R both are used in Data.",
    "python has a bunch of libraries that help data-scientists."
]

# Tokenize the documents into words
tokenized_documents = [doc.split() for doc in documents]
print(f"sample :{tokenized_documents[0]}")
print(f"tokenized documents length: {len(tokenized_documents)}")
# Train a Word2Vec model on the tokenized documents
model = Word2Vec(tokenized_documents, vector_size=10, window=3, min_count=1, workers=4)
print(f"model: {model}")
# Calculate cosine similarity using word embeddings
embedding_similarity = cosine_similarity([model.wv['python']], [model.wv['R']])
print(f"Cosine Similarity using Word Embeddings: {embedding_similarity[0][0]}")
```
